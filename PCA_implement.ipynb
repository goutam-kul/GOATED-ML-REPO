# Principle Component Analysis (PCA)

- PCA is a common Dimensionality Reduction technique 

```python
# data wrangling 
import numpy as np
import pandas as pd

# dataset
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA
# visualization
import seaborn as sns 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# make swiss role dataset from sklearn class 
X, color = make_swiss_roll(n_samples=1000, random_state=42, hole=True)
```

```python
X[1]
```

```python
color[1]
```

# Principle Components 

```python
# Define the pca with components 2 
pca = PCA(n_components=2)

# compute pca for data X
X_pca = pca.fit_transform(X)
```

```python
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.inferno)
ax1.set_title("Original 3D Swiss roll")

# PCA components
ax2 = fig.add_subplot(132)
ax2.scatter(X[:, 0], X[:, 1], c=color, cmap=plt.cm.viridis)
ax2.set_title("First Two Original Dimensions")

# PCA result
ax3 = fig.add_subplot(133)
ax3.scatter(X_pca[:, 0], X_pca[:, 1], c=color, cmap=plt.cm.viridis)
ax3.set_title("2D PCA Projection")

plt.tight_layout()
plt.show()
```

- You can see the difference between just using first two dimensions
- and when we take projection of the 3D data using PCA
- the PCA captures the maximum variance in the dataset while just using two dimensions not so.
- PCA captures conserves most amount of information as we can see.
